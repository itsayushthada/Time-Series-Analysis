```r
WhiteNoise <- function(len, mean=0, sd=1){
    return(ts(rnorm(len, mean = mean, sd = sd)))
}
```

## Note: On inversion of MA(q) ie. Moving Average Process we receive AR process. MA processes are always statonary but this is not a case with AR Processes.

```r
set.seed(2016); 
N  <- 1000; 
phi <- .4;

Z <- rnorm(N,0,1); X=NULL; X[1] = Z[1];

for (t in 2:N) {
    X[t] <- Z[t] + phi*X[t-1] ;
}
X.ts <- ts(X)

par(mfrow=c(2,1))
plot(X.ts,main="AR(1) Time Series on White Noise, phi=.4")
X.acf <- acf(X.ts, main="AR(1) Time Series on White Noise, phi=.4")
```

```r
# On Phi Value 1 it behaves just as Random Noise. Also as we increase the value of phi decay become slower and slower.

set.seed(2016); 
N  <- 1000; 
phi <- 1;

Z <- rnorm(N,0,1); X=NULL; X[1] = Z[1];

for (t in 2:N) {
    X[t] <- Z[t] + phi*X[t-1] ;
}
X.ts <- ts(X)

par(mfrow=c(2,1))
plot(X.ts,main="AR(1) Time Series on White Noise, phi=1")
X.acf <- acf(X.ts, main="AR(1) Time Series on White Noise, phi=1")
```

# Auto Regressive Process

```r
AutoRegProcess <- function(phis, len=100, process=1, mean=0, sd=1){
    "
    This function generates moving average process.
    theta <- fraction of previous noise affeacting current result.
    "
    order <- length(phis)
    len <- 100 + order
    
    AR<- matrix(rep(0,len*process), nrow = len, ncol = process)

    AR[1:order,] <- matrix(rnorm(order*process, mean, sd), nrow = order, ncol = process)

    if(process>1){
        for(i in (order+1):len){
             AR[i,] <- rnorm(process, mean, sd) + colSums(phis * AR[(i-order):(i-1),]) 
        }
    }
    else{
        for(i in (order+1):len){
             AR[i,] <- rnorm(process, mean, sd) + sum(phis * AR[(i-order):(i-1),]) 
        }
    }

    return(ts(AR[(order+1):len,]))
}
```

```r
# Test the function

AR <- AutoRegProcess(phis = c(0.9, 0.1), len = 500, process = 3)

options(repr.plot.width=10, repr.plot.height=6)
matplot(AR, type = 'l', 
        col=c("red", "green", "blue", "brown", "orange"), 
        lty = c(1, 1, 1, 1, 1), xlab = "Days", ylab="Variation")

legend(x = "bottomleft", 
       legend = c("I AR", "II AR", "III AR", "IV AR", "V AR"), 
       col = c("red", "green", "blue", "brown", "orange"), 
       lty = rep(1,5), 
       cex=0.7)
```

# AR(2) process comparsion with White Noise

```r
options(repr.plot.width=10, repr.plot.height=12)
N <- 500
theta <-c(0.6, 0.3)

ARq <- AutoRegProcess(phis = theta, len = N)
WN <- WhiteNoise(N)

par(mfrow=c(4, 1))
plot(ARq, type = 'l',col=c("red"), xlab = "Days", ylab="Time Utilized")
(acf(ARq, col="red"))
plot(WN, type = 'l',col=c("blue"), xlab = "Days", ylab="Time Utilized")
(acf(WN, col="blue"))
```
