```r
WhiteNoise <- function(len, mean=0, sd=1){
    return(ts(rnorm(len, mean = mean, sd = sd)))
}
```

# Moving Average Processes

```r
MovingAvgProcess <- function(theta, len=100, process=1, mean=0, sd=1){
    "
    This function generates moving average process.
    theta <- fraction of previous noise affeacting current result.
    "
    theta <- theta/sum(theta)
    theta <- c(theta, 1)
    ord <- length(theta) - 1
    len <- ord + len
    X <- matrix(rnorm(len*process, mean = mean, sd = sd), nrow = len, ncol = process)
    MA <- matrix(, nrow = len-ord, ncol = process)

    if(process > 1){
        for(i in (ord+1):len){
            MA[i-ord, 1:process] <- colSums(X[as.integer(i-ord):i, 1:process] * theta)
        }
    }
    else{
        for(i in (ord+1):len){
            MA[i-ord, 1:process] <- sum(X[as.integer(i-ord):i, 1:process] * theta)
        }
    }

    return(ts(MA))
}
```

### MA(3) process comparsion with White Noise

```r
options(repr.plot.width=10, repr.plot.height=12)
N <- 500
theta <-c(0.6, 0.3, 0.1)

MAq <- MovingAvgProcess(len = N, theta = theta)
WN <- WhiteNoise(N)

par(mfrow=c(4, 1))
plot(MAq, type = 'l',col=c("red"), xlab = "Days", ylab="Time Utilized")
(acf(MAq, col="red"))
plot(WN, type = 'l',col=c("blue"), xlab = "Days", ylab="Time Utilized")
(acf(WN, col="blue"))

# MA3 series has some structure as compared to pure white noise. The structure will get stronger and stronger
# once we start considering more values for mean operation along with appropirate parametrs.
```

```r
# We cansimulate it using the R functions as well.

options(repr.plot.width=10, repr.plot.height=8)
MA3 <- arima.sim(n=N, model=list(ma=theta))

par(mfrow=c(2, 1))
plot(MA3, type = 'l',col=c("red"), xlab = "Days", ylab="Time Utilized")
(acf(MA3, col="red"))
```

### MA(8) process comparsion with White Noise

```r
options(repr.plot.width=10, repr.plot.height=12)
N <- 500

theta <- c(100, 80, 60, 75, 40, 20, 15, 10)
theta <- (theta - mean(theta))/(max(theta) - min(theta))
MAq <- MovingAvgProcess(len = N, theta = theta)
WN <- WhiteNoise(N)

par(mfrow=c(4, 1))
plot(MAq, type = 'l',col=c("red"), xlab = "Days", ylab="Time Utilized")
(acf(MAq, col="red"))
plot(WN, type = 'l',col=c("blue"), xlab = "Days", ylab="Time Utilized")
(acf(WN, col="blue"))

# In the AutoCorrealtion Function graph of MA(8) process peaks are there. It suggest the weak stationrity.
```

```r
# We cansimulate it using the R functions as well.

options(repr.plot.width=10, repr.plot.height=8)
MA3 <- arima.sim(n=N, model=list(ma=theta))

par(mfrow=c(2, 1))
plot(MA3, type = 'l',col=c("red"), xlab = "Days", ylab="Time Utilized")
(acf(MA3, col="red"))
```

```r
AutoCovariance <- function(param, lag=0, sd=1){
    "
    param <- It is the list of cofficeints which are multiplied with the value of white noise ina time slice.
    lag <- It is the time sepraion between the value in time series whse ACF is to be caclulated.
    sd <- Standard deviation of the underlying white noise.
    "
    res <- 0
    q <- length(param)
    for(i in 1:(q-lag)){
        res <- res + (param[i]*param[i+lag])
    }
    return(sd*sd*res)
}
```

```r
AutoCoRelation <- function(param, lag=0, sd=1){
    "
    param <- It is the list of cofficeints which are multiplied with the value of white noise ina time slice.
    lag <- It is the time sepraion between the value in time series whse ACF is to be caclulated.
    sd <- Standard deviation of the underlying white noise.
    "
    return(AutoCovariance(param, lag, sd)/AutoCovariance(param, 0, sd))
}
```
