# Random Walk

```r
options(repr.plot.width=10, repr.plot.height=5)
```

```r
NonStatRandomWalk <- function(len=100, process=1){
    "
    This function generates Non Stationary Time Series.
    "
    X = matrix(, nrow = len, ncol = process)

    X[1, 1:process] <- rep(30, process)
    for(i in 2:len){
        X[i, 1:process] <- X[i-1, 1:process] + rnorm(process)
    }

    return(ts(X))
}

# This pocess is not stationay because both variance and expectation is dependet on time.
```

```r
X <- NonStatRandomWalk(600, 5)

matplot(X, type = 'l', 
        col=c("red", "green", "blue", "brown", "orange"), 
        lty = c(1, 1, 1, 1, 1), xlab = "Days", ylab="Forex Price")

legend(x = "bottomleft", 
       legend = c("I RW", "II RW", "III RW", "IV RW", "V RW"), 
       col = c("red", "green", "blue", "brown", "orange"), 
       lty = rep(1,5), 
       cex=0.7)
```

```r
# We ususally calcualte Autocorrelation for stationary time series.
# But let's try it for the first time series.

plot(acf(X[1:600,1]))

# High correlation shows there is no stationarity.
```

```r
StatRandomWalk <- function(len=100, process=1, lag=1){
    "
    This function generates Stationary Time Series.
    Lower the value of lag less autocorrelation will be there.
    "
    X <- NonStatRandomWalk(len = len+lag, process = process)
    X <- as.matrix.data.frame(diff(X, lag = lag))
    return(ts(X))
}
```

```r
Y <- StatRandomWalk(600, 2, lag = 1)

matplot(Y, type = 'l', 
        col=c("red", "green"), 
        lty = rep(1,2), 
        xlab = "Days", 
        ylab="Forex Price")

legend(x = "bottomleft", 
       legend = c("I RW", "II RW", "III RW", "IV RW", "V RW"), 
       col = c("red", "green"), 
       lty = rep(1,2), 
       cex=0.7)
```

```r
plot(acf(Y[1:600,1]))
plot(acf(Y[1:600,2]))

# Low correlation shows there is no/weak stationarity.
```

## Moving Average Processes

```r
MovingAvgProcess <- function(theta, len=100, process=1){
    "
    This function generates moving average process.
    theta <- fraction of previous noise affeacting current result.
    "
    theta <- theta/sum(theta)
    theta <- c(theta, 1)
    ord <- length(theta) - 1
    len <- ord + len
    X <- matrix(rnorm(len*process), nrow = len, ncol = process)
    MA <- matrix(, nrow = len-ord, ncol = process)

    if(process > 1){
        for(i in (ord+1):len){
            MA[i-ord, 1:process] <- colSums(X[as.integer(i-ord):i, 1:process] * theta)
        }
    }
    else{
        for(i in (ord+1):len){
            MA[i-ord, 1:process] <- sum(X[as.integer(i-ord):i, 1:process] * theta)
        }
    }

    return(ts(MA))
}
```

```r
Z <- MovingAvgProcess(len = 1000, theta = c(0.7, 0.3), process = 2)

matplot(Z, type = 'l', 
        col=c("red", "green"), 
        lty = rep(1,2), 
        xlab = "Days", 
        ylab="Time Utilized")

legend(x = "bottomleft", 
       legend = c("I RW", "II RW", "III RW", "IV RW", "V RW"), 
       col = c("red", "green"), 
       lty = rep(1,2), 
       cex=0.7)
```

```r
acf(Z[1:1000, 1], main="Correlogram of I moving average process of order 2")
acf(Z[1:1000, 2], main = "Correlogram of II moving average process of order 2")
```
